```python
# @title Year vs count()

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  counted = (series['Year']
                .value_counts()
              .reset_index(name='counts')
              .rename({'index': 'Year'}, axis=1)
              .sort_values('Year', ascending=True))
  xs = counted['Year']
  ys = counted['counts']
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Year', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Period')):
  _plot_series(series, series_name, i)
  fig.legend(title='Period', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Year')
_ = plt.ylabel('count()')
```

```python
import pandas as pd
import numpy as np

def decompose_rows(df):
    # Ensure all relevant columns are strings (or NaNs)
    df['Crop'] = df['Crop'].astype(str)
    df['Country'] = df['Country'].astype(str)
    df['Period'] = df['Period'].astype(str)
    df['mycotoxin'] = df['mycotoxin'].astype(str)

    # Split 'Crop' on whitespace
    df['Crop'] = df['Crop'].str.split(r'\s+')
    df['Crop'] = df['Crop'].apply(lambda x: [i.strip().lower() for i in x] if isinstance(x, list) else [str(x).strip().lower()])

    # Split 'Country' on newline (if needed)
    df['Country'] = df['Country'].str.split(r'\n')
    df['Country'] = df['Country'].apply(lambda x: [i.strip().lower() for i in x] if isinstance(x, list) else [str(x).strip().lower()])

    # Split 'Period' on 'and' or newline
    df['Period'] = df['Period'].str.replace('\nand', ' and ', regex=True)
    df['Period'] = df['Period'].str.split(r'\s*and\s*')
    df['Period'] = df['Period'].apply(lambda x: [i.strip().lower() for i in x] if isinstance(x, list) else [str(x).strip().lower()])

    # Split 'mycotoxin' on newline, handling NaN or empty values
    df['mycotoxin'] = df['mycotoxin'].replace(['nan', '', 'NaN'], np.nan)

    df['mycotoxin'] = df['mycotoxin'].dropna().astype(str).str.split(r'\n')
    df['mycotoxin'] = df['mycotoxin'].apply(lambda x: [i.strip().lower() for i in x if i.strip()] if isinstance(x, list) else [str(x).strip().lower()])



    # Explode all the columns
    df = df.explode('Crop')
    df = df.explode('Country')
    df = df.explode('Period')
    df = df.explode('mycotoxin')

    # Reset index for clarity
    df = df.reset_index(drop=True)

    return df

# Example usage:
# df = pd.read_csv('your_data.csv')  # or however you get your dataframe
df_decomposed_cc = decompose_rows(df.copy())
# 3. Fix inconsistent country names
myco_name_map = {
    'fumonisins': "fumonisin",
    'ochratoxin a': 'ochratoxin',
    'deoxinivalenol': 'deoxynivalenol',
    'fumonisin zearalenone': 'fumonisin',
    'fumonisin zearalenone': 'zearalenone'
}
df_decomposed_cc['mycotoxin'] = df_decomposed_cc['mycotoxin'].replace(myco_name_map)
print(df_decomposed_cc.info())
print(df_decomposed_cc.head())
```

```python
!pip install -U kaleido
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# 3. Fix inconsistent country names
country_name_map = {
    'ivory coast': "cote d'ivoire",
    'drc': 'democratic republic of the congo',
    'S. Sudan': 'sudan',
    'cameroun': 'cameroon'
}
df_decomposed_cc['Country'] = df_decomposed_cc['Country'].replace(country_name_map)

# Step 1: Normalize names
df_decomposed_cc['Country'] = df_decomposed_cc['Country'].str.strip().str.title()
df_decomposed_cc['Year'] = df_decomposed_cc['Year'].astype(str)

# Step 2: Group data — count records per Country per Year
country_year_counts = (
    df_decomposed_cc
    .groupby(['Year', 'Country'])
    .size()
    .reset_index(name='Count')
)

# *** Include 2015 even if it has no data ***
all_years = ['2015'] + sorted(country_year_counts['Year'].unique())
all_years= sorted(all_years)
country_year_counts['Year'] = pd.Categorical(country_year_counts['Year'], categories=all_years, ordered=True)

# Step 3: Setup subplot grid
n_years = len(all_years)
cols = 3
rows = (n_years + cols - 1) // cols

fig = make_subplots(
    rows=rows, cols=cols,
    subplot_titles=all_years,  # Use all_years for subplot titles
    specs=[[{"type": "choropleth"}] * cols for _ in range(rows)],
    horizontal_spacing=0.01,
    vertical_spacing=0.01
)

# Step 4: Add traces per subplot
for i, year in enumerate(all_years):
    row = i // cols + 1
    col = i % cols + 1
    data_for_year = country_year_counts[country_year_counts['Year'] == year]

    # Only the last trace should show the shared colorbar
    show_colorbar = (i == len(all_years) - 1)

    choropleth = go.Choropleth(
        locations=data_for_year['Country'],
        locationmode='country names',
        z=data_for_year['Count'],
        colorscale='Viridis',
        zmin=country_year_counts['Count'].min(),
        zmax=country_year_counts['Count'].max(),
        colorbar=dict(title='Record Count'),
        coloraxis='coloraxis',
        showscale=show_colorbar
    )

    fig.add_trace(choropleth, row=row, col=col)

# Step 5: Update layout
fig.update_layout(
    title_text="Country-Level Record Counts per Year (Africa)",
    height=220 * rows,
    coloraxis=dict(colorscale='Viridis'),
    margin=dict(l=10, r=10, t=80, b=10)
)

# Apply Africa scope to all subplots
for i in range(n_years):
    fig.update_geos(
        scope='africa',
        showcoastlines=False,
        showframe=False,
        row=(i // cols + 1),
        col=(i % cols + 1)
    )

fig.show()
fig.write_image("country_record_counts_africa.png", scale=2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Normalize country and crop names
df_decomposed_cc['Country'] = df_decomposed_cc['Country'].str.lower().str.strip()
df_decomposed_cc['Crop'] = df_decomposed_cc['Crop'].str.lower().str.strip()

# 2. Define country groups (regional stack)
country_groups = {
    'West Africa': ['nigeria', 'ghana', 'ivory coast', 'gambia', 'senegal', 'togo'],
    'East Africa': ['kenya', 'ethiopia', 'tanzania', 'uganda', 'burundi', 'rwanda'],
    'Southern Africa': ['south africa', 'zimbabwe', 'malawi', 'namibia'],
    'North Africa': ['tunisia', 'morocco', 'algeria'],
    'Central Africa': ['cameroun', 'drc', 'zambia'],
    'Sahel': ['mali', 'niger', 'south sudan']
}

# 3. Fix inconsistent country names
country_name_map = {
    'ivory coast': "cote d'ivoire",
    'drc': 'democratic republic of the congo',
    'south sudan': 's. sudan',
    'cameroun': 'cameroon'
}
df_decomposed_cc['Country'] = df_decomposed_cc['Country'].replace(country_name_map)

# 4. Get list of unique crops for fixed color palette
all_crops = sorted(df_decomposed_cc['Crop'].unique())
palette = sns.color_palette("Set2", len(all_crops))
crop_palette = dict(zip(all_crops, palette))

# 5. Create subplot grid
fig, axes = plt.subplots(3, 2, figsize=(16, 12), constrained_layout=True)

# 6. Loop through each region and plot timeline
for (region, countries), ax in zip(country_groups.items(), axes.flatten()):
    subset = df_decomposed_cc[df_decomposed_cc['Country'].isin(countries)]

    # Group and count by region, crop, and year
    region_grouped = (
        subset.groupby(['Year', 'Crop'])
        .size()
        .reset_index(name='Count')
    )

    # Line plot per crop over years
    sns.lineplot(
        data=region_grouped,
        x='Year', y='Count',
        hue='Crop',
        ax=ax,
        hue_order=all_crops,
        palette=crop_palette,
        marker='o'
    )

    ax.set_title(f"{region}")
    ax.set_ylabel("Crop Count")
    ax.set_xlabel("Year")
    ax.tick_params(axis='x', rotation=45)
    ax.legend().remove()

# 7. Global legend
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=len(all_crops), bbox_to_anchor=(0.5, 1.02))

# 8. Title and save
plt.suptitle('Crop Occurrences Over Time by Region and Crop Type', fontsize=16, y=1.08)
plt.savefig("region_crop_timeline.png", dpi=300, bbox_inches='tight')
plt.show()
```
